```python
def compare_quantity_total(df1, df2):
    difference_between_years = 0
    if df1["Date year"][0] == df2["Date year"][0]:
        difference_between_years = 0
    else:
        if df1["Date year"][0] == df2["Date year"][0] + 1:
            difference_between_years = -1
        elif df1["Date year"][0] + 1 == df2["Date year"][0]:
            difference_between_years = 1
        else:
            difference_between_years = 2
    
    diff = False
    if difference_between_years == 0:
        if df1["Quantity SY"][0] == df2["Quantity SY"][0] and df1["Quantity PY"][0] == df2["Quantity PY"][0]:
            diff = True
    
    if difference_between_years == 1:
        if df1["Quantity SY"][0] == df2["Quantity PY"][0]:
            diff = True
    
    if difference_between_years == -1:
        if df1["Quantity PY"][0] == df2["Quantity SY"][0]:
            diff = True

    return diff, difference_between_years
```

```python
def compare_value_total(df1, df2):
    difference_between_years = 0
    if df1["Date year"][0] == df2["Date year"][0]:
        difference_between_years = 0
    else:
        if df1["Date year"][0] == df2["Date year"][0] + 1:
            difference_between_years = -1
        elif df1["Date year"][0] + 1 == df2["Date year"][0]:
            difference_between_years = 1
        else:
            difference_between_years = 2
    
    s = df1["Value SY"][0]
  
    if isinstance(s, str):
        if not s[0].isdigit(): 
            df1["Value SY"][0] =  float(s[1:])
        elif not s[-1].isdigit():
            df1["Value SY"][0] =  float(s[:-2])

    s = df1["Value PY"][0]
    if isinstance(s, str):
        if not s[0].isdigit(): 
            df1["Value PY"][0] =  float(s[1:])
        elif not s[-1].isdigit():
            df1["Value PY"][0] =  float(s[:-2])

    s = df2["Value SY"][0]
    if isinstance(s, str):
        if not s[0].isdigit(): 
            df2["Value SY"][0] =  float(s[1:])
        elif not s[-1].isdigit():
            df2["Value SY"][0] =  float(s[:-2])

    s = df2["Value PY"][0]
    if isinstance(s, str):
        if not s[0].isdigit(): 
            df2["Value PY"][0] =  float(s[1:])
        elif not s[-1].isdigit():
            df2["Value PY"][0] =  float(s[:-2])
    
    
    diff = False
    diff_perc = 0
    division1 = 0
    division2 = 0
    diff_division = 0
    if difference_between_years == 0: 
        diff_perc = abs(int(df1["Value SY"][0]) - int(df2["Value SY"][0])) + abs(int(df1["Value PY"][0]) - int(df2["Value PY"][0])) 
        division1 = int(df1["Value SY"][0]) / int(df1["Value PY"][0])
        division2 = int(df2["Value SY"][0]) / int(df2["Value PY"][0])
        diff_division = division1 - division2
        if df1["Value SY"][0] == df2["Value SY"][0] and df1["Value PY"][0] == df2["Value PY"][0]:
            diff = True
    
    if difference_between_years == 1:
        diff_perc = abs(int(df1["Value SY"][0]) - int(df2["Value PY"][0]))
        if df1["Value SY"][0] == df2["Value PY"][0]:
            diff = True
    
    if difference_between_years == -1:
        diff_perc = abs(int(df1["Value PY"][0]) - int(df2["Value SY"][0]))
        if df1["Value PY"][0] == df2["Value SY"][0]:
            diff = True



    
    return diff, difference_between_years, diff_perc, division1, division2, diff_division
```

```python
def compare_quantity_country(df1, df2, diff_years):
    if diff_years == 1:
        df1, df2 = df2, df1
    diff = True
    if len(pd.unique(df1["Country"])) != len(pd.unique(df2["Country"])):
        return False
    for country in pd.unique(df1["Country"]):

        q1_sy = pd.DataFrame(df1.loc[df1["Country"] == country]["Quantity SY"])
        q1_py = pd.DataFrame(df1.loc[df1["Country"] == country]["Quantity PY"])
        q2_sy = pd.DataFrame(df2.loc[df2["Country"] == country]["Quantity SY"])
        q2_py = pd.DataFrame(df2.loc[df2["Country"] == country]["Quantity PY"])

        q1_sy = q1_sy.reset_index()
        q1_py = q1_py.reset_index()
        q2_sy = q2_sy.reset_index()
        q2_py = q2_py.reset_index()
        

        q1_py = q1_py.iloc[0, 1]
        q1_sy = q1_sy.iloc[0, 1]
        q2_py = q2_py.iloc[0, 1]
        q2_sy = q2_sy.iloc[0, 1]
        
        
        if diff_years == 0:
            if ((q1_py != q2_py) or (q1_sy != q2_sy)):
                diff = False

        if diff_years == -1 or diff_years == 1:
            if (q1_py != q2_sy):
                diff = False
        
    return diff
```

```python
def compare_quantity_month(df1, df2, diff_years):

    if diff_years == 1:
        df1, df2 = df2, df1
    diff = True
    df1 = df1.dropna()
    df2 = df2.dropna()
    if len(df1) != len(df2):
        return False
    for month in pd.unique(df1["Month short"]):
        df1_m = df1[df1["Month short"] == month]
        df2_m = df2[df2["Month short"] == month]
        for week in pd.unique(df1_m["Date week"]):
            
            q1_sy = pd.DataFrame(df1_m.loc[df1_m["Date week"] == week]["Quantity SY"])
            q1_py = pd.DataFrame(df1_m.loc[df1_m["Date week"] == week]["Quantity PY"])
            q2_sy = pd.DataFrame(df2_m.loc[df2_m["Date week"] == week]["Quantity SY"])
            q2_py = pd.DataFrame(df2_m.loc[df2_m["Date week"] == week]["Quantity PY"])

            q1_sy = q1_sy.reset_index()
            q1_py = q1_py.reset_index()
            q2_sy = q2_sy.reset_index()
            q2_py = q2_py.reset_index()
        
            if len(q1_py) == 0 or len(q2_py) == 0 or len(q1_sy) == 0 or len(q2_sy) == 0:
                continue
            q1_py = q1_py.iloc[0, 1]
            q1_sy = q1_sy.iloc[0, 1]
            q2_py = q2_py.iloc[0, 1]
            q2_sy = q2_sy.iloc[0, 1]
            
            
            if diff_years == 0:
                if ((q1_py != q2_py) or (q1_sy != q2_sy)):
                    diff = False
                

            if diff_years == -1 or diff_years == 1:
                if (q1_py != q2_sy):
                    diff = False
        
    return diff
```

```python
def compare(pair_name):
    predicion = {}
    vis = []
    status = []
    sy_vs_py_q_1, sy_vs_py_q_2, country_q_1, country_q_2, customer_q_1, customer_q_2, month_q_1, month_q_2, veggie_q_1, veggie_q_2, sy_vs_py_v_1, sy_vs_py_v_2, country_v_1, country_v_2, customer_v_1, customer_v_2, month_v_1, month_v_2, veggie_v_1, veggie_v_2 = get_data(pair_name)
    
    diff, diff_years = compare_quantity_total(sy_vs_py_q_1, sy_vs_py_q_2)
    vis.append("Quantity total")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")
    diff = compare_quantity_country(country_q_1, country_q_2, diff_years=diff_years)
    vis.append("Sales quantity by country")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")
    diff = compare_quantity_customer(customer_q_1, customer_q_2, diff_years=diff_years)

    vis.append("Sales quantity by customer")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")
    diff = compare_quantity_veggie(veggie_q_1, veggie_q_2, diff_years=diff_years)

    vis.append("Sales quantity by veggie")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")
    diff = compare_quantity_month(month_q_1, month_q_2, diff_years=diff_years)

    vis.append("Sales quantity")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")
    diff, _, _, _ , _, _= compare_value_total(sy_vs_py_v_1, sy_vs_py_v_2)

    vis.append("Value total")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")
    diff, _, _, _, _ = compare_value_country(country_v_1, country_v_2, diff_years=diff_years)
    vis.append("Sales value by country")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")
    diff, _, _, _ , _= compare_value_customer(customer_v_1, customer_v_2, diff_years=diff_years)
    vis.append("Sales value by customer")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")

    diff, _, _, _, _ = compare_value_veggie(veggie_v_1, veggie_v_2, diff_years=diff_years)
    vis.append("Sales value by veggie")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")
    diff, _, _, _ , _= compare_value_month(month_v_1, month_v_2, diff_years=diff_years)
    vis.append("Sales value")
    if diff:
        status.append("Match")
    else:
        status.append("Not match")

    predicion["VisualObject"] = vis
    predicion["Status"] = status
    return predicion
```
